## chapter02 자료형
### 01.자료형과 문자열

```python
# python의 자료형 : 문자열, 숫자, 불
print('문자열의 데이터 타입 : ', type('문자열'))
print('숫자의 데이터 타입 : ', type(10), type(10.0))
print('불의 데이터 타입 : ', type(True))
```

```python
#"" ''를 출력하고싶을때
print('"안녕하세요!"')
print("'안녕하세요!'")
# 이스케이프 문자를 활용하여 문자열 만들기 : \" -> "
print("\"안녕하세요!\"")
# \t : 탭, \n :다음라인
print("\"안녕\t하세요\n!\"")
```
